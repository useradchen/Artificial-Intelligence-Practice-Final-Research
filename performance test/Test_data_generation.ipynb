```python
#由gemini 2.5 pro 輔助撰寫
import sys
# 假設您的 index-tts 儲存庫被複製到了 /content/index-tts/
project_root = '/content/index-tts'
if project_root not in sys.path:
    sys.path.append(project_root)

# 現在 Python 應該能夠找到 indextts 模組了
from indextts.infer import IndexTTS

# ------------ 您的其餘程式碼 ------------

# 請確保您的模型和配置文件路徑是正確的
# 根據您之前的操作，模型可能在 'models/' 而非 'checkpoints/'
# 且配置文件可能是 'models/config.json'
# 請根據您下載模型時的實際路徑進行調整
# model_directory = "models" # 例如，如果您的 .pth 文件都在這裡
# config_file_path = "models/config.json" # 例如，如果這是您的配置文件

# 暫時使用您提供的路徑，但請您確認這些路徑是否正確
model_directory = "checkpoints"
config_file_path = "checkpoints/config.yaml"

# 確保 "checkpoints" 目錄和 "config.yaml" 存在於 /content/index-tts/checkpoints/
# 如果不存在，您需要創建它們並放入相應文件，或者修改路徑指向正確的模型和配置文件

# 確保參考音訊存在
# 例如，上傳到 /content/index-tts/reference_voice.wav
reference_voice_path = "girl-ch.wav"

# 定義輸出路徑
output_audio_path = "generated_audio_colab.wav"

text_to_synthesize = "月光灑落靜謐湖面，微波蕩漾，映照夢境，彷彿時光停駐，訴說永恆的詩意。"

try:
    # 根據 IndexTTS 類的實際參數初始化
    # 您可能需要檢查 indextts/infer.py 中 IndexTTS 類的 __init__ 方法
    # 來確定 model_dir 和 cfg_path 是否是正確的參數名和格式
    tts = IndexTTS(model_dir=model_directory, cfg_path=config_file_path)

    print(f"正在使用參考音訊: {reference_voice_path}")
    print(f"合成文本: {text_to_synthesize}")
    print(f"輸出到: {output_audio_path}")

    # 執行推斷
    tts.infer(reference_voice_path, text_to_synthesize, output_audio_path)

    print(f"語音合成完成！輸出文件應為: {output_audio_path}")

    # 播放生成的音訊
    import IPython.display as ipd
    display(ipd.Audio(output_audio_path))

except FileNotFoundError as e:
    print(f"文件未找到錯誤: {e}")
    print("請確保您的模型文件、配置文件和參考音訊的路徑是正確的，並且這些文件都已上傳到 Colab。")
except Exception as e:
    print(f"發生錯誤: {e}")
    print("請檢查 IndexTTS 類的初始化參數以及 infer 方法的參數是否正確。")
```
